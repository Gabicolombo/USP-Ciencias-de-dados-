```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Importando os dados do csv

```python
dataset = pd.read_csv("hcvdat0.csv")

display(dataset)
```

```python
df = pd.DataFrame(dataset)

# Contabilizar os NAs em cada coluna
df_nas = df.isna().sum()

print("Valores ausentes por coluna:")
print(df_nas)
```

```python
df.describe()
```

```python
def glimpse(df):
    for col in df.columns:
        print(f"$ {col} <{df[col].dtype}>")

glimpse(df)
```

```python
plt.figure(figsize=(10, 8))

df.boxplot()

plt.title('Dados do dataset de hepatite c')
plt.ylabel('Valores')

plt.xticks(rotation=45)  
plt.show()
```

```python
df_category = df.loc[:, ['Category', "ALB", "ALP", "ALT", "AST", "BIL", "CHE", "CHOL", "CREA", "GGT", "PROT"]]
df_category
```

```python
# Agrupando os dados por categoria e calculando a mÃ©dia das variÃ¡veis numÃ©ricas
grouped_df = df_category.groupby('Category').mean()

# ConfiguraÃ§Ãµes de plotagem
plt.figure(figsize=(10, 6))

# Plotando as barras para cada variÃ¡vel numÃ©rica
for col in grouped_df.columns:
    plt.bar(grouped_df.index, grouped_df[col], label=col)

# Adicionando tÃ­tulo e rÃ³tulos aos eixos
plt.title('MÃ©dia das variÃ¡veis por categoria')
plt.xlabel('Categoria')
plt.ylabel('Valor MÃ©dio')
plt.legend()

# Exibindo o grÃ¡fico
plt.tight_layout()
plt.show()
```
