```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Importando os dados do csv

```python
dataset = pd.read_csv("hcvdat0.csv")

display(dataset)
```

```python
df = pd.DataFrame(dataset)

# Contabilizar os NAs em cada coluna
df_nas = df.isna().sum()

print("Valores ausentes por coluna:")
print(df_nas)
```

```python
df.describe()
```

```python
def glimpse(df):
    for col in df.columns:
        print(f"$ {col} <{df[col].dtype}>")

glimpse(df)
```

```python
def plot(df):
    plt.figure(figsize=(10, 8))

    df.boxplot()

    plt.title('Dados do dataset de hepatite c')
    plt.ylabel('Valores')

    plt.xticks(rotation=45)  
    plt.show()
```

```python
plot(df)
```

```python
df_category = df.loc[:, ['Category', "ALB", "ALP", "ALT", "AST", "BIL", "CHE", "CHOL", "CREA", "GGT", "PROT"]]
df_category
```

```python
def plot_mean(df):
    grouped_df = df.groupby('Category').mean()
    plt.figure(figsize=(10, 6))

    for col in grouped_df.columns:
        plt.bar(grouped_df.index, grouped_df[col], label=col)

    plt.title('MÃ©dia das variÃ¡veis por categoria')
    plt.xlabel('Categoria')
    plt.ylabel('Valor MÃ©dio')
    plt.legend()

    plt.tight_layout()
    plt.show()

def plot_mean2(df):
    df = df.groupby('Category').mean()
    plt.figure(figsize=(10, 6))

    df.plot(kind='bar', figsize=(10, 6))
    plt.title('MÃ©dia dos Valores por Categoria')
    plt.xlabel('Categoria')
    plt.ylabel('MÃ©dia dos Valores')
    plt.xticks(rotation=0)  # Rotaciona os rÃ³tulos do eixo x para facilitar a leitura
    plt.legend(title='Coluna')
    plt.show()

def plot_sex_category(df):
    sex_count = df.groupby(['Category', 'Sex']).size().unstack()

    sex_count.plot(kind='bar', stacked=True, figsize=(10, 6))
    plt.title('Quantidade de "m" e "f" por Categoria')
    plt.xlabel('Categoria')
    plt.ylabel('Quantidade')
    plt.xticks(rotation=0)  
    plt.legend(title='Sexo')
    plt.show()
```

```python
plot_mean2(df_category)
```

```python
df_age_category_1 = df.loc[:, ['Age', 'Category']]
plot_mean2(df_age_category_1)
```

```python
plot_sex_category(df)
```
